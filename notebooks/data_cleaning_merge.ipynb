Mergin tables and exporting two tables performance_fact and employee_master

```python

# --- 1. Prepare Lookup Tables for Merging ---
df_satisfied_lookup = satisfaction_levelDF.rename(
    columns={'satisfactionid': 'id', 'satisfactionlevel': 'satisfaction_text'}
)[['id', 'satisfaction_text']]

df_rating_lookup = rating_levelDF.rename(
    columns={'ratingid': 'id', 'ratinglevel': 'rating_text'}
)[['id', 'rating_text']]

# --- 2. Create the Enriched Performance Fact Table (performance_fact.csv) ---

print("Starting: Creating enriched Performance Fact Table...")

df_fact = performanceDF.copy()

# --- A. Merge Satisfaction Levels (3 times) ---

# 1. Environment Satisfaction
df_fact = pd.merge(
    df_fact,
    df_satisfied_lookup.rename(columns={'satisfaction_text': 'envsatisfaction_text'}),
    left_on='environmentsatisfaction',
    right_on='id',
    how='left'
)
df_fact.drop(columns=['id'], inplace=True, errors='ignore')

# 2. Job Satisfaction
df_fact = pd.merge(
    df_fact,
    df_satisfied_lookup.rename(columns={'satisfaction_text': 'jobsatisfaction_text'}),
    left_on='jobsatisfaction',
    right_on='id',
    how='left'
)
df_fact.drop(columns=['id'], inplace=True, errors='ignore')

# 3. Relationship Satisfaction
df_fact = pd.merge(
    df_fact,
    df_satisfied_lookup.rename(columns={'satisfaction_text': 'relationshipsatisfaction_text'}),
    left_on='relationshipsatisfaction',
    right_on='id',
    how='left'
)
df_fact.drop(columns=['id'], inplace=True, errors='ignore')

# --- B. Merge Rating Levels (2 times) ---
# 1. Self Rating
df_fact = pd.merge(
    df_fact,
    df_rating_lookup.rename(columns={'rating_text': 'selfrating_text'}),
    left_on='selfrating',
    right_on='id',
    how='left'
)
df_fact.drop(columns=['id'], inplace=True, errors='ignore')

# 2. Manager Rating
df_fact = pd.merge(
    df_fact,
    df_rating_lookup.rename(columns={'rating_text': 'managerrating_text'}),
    left_on='managerrating',
    right_on='id',
    how='left'
)
df_fact.drop(columns=['id'], inplace=True, errors='ignore')

# --- 3. Feature Engineering ---
df_fact['reviewdate'] = pd.to_datetime(df_fact['reviewdate'], errors='coerce')
df_fact.dropna(subset=['reviewdate'], inplace=True)

# Add Review Year
df_fact['reviewyear'] = df_fact['reviewdate'].dt.year

# Sort for lag calculations
df_fact.sort_values(by=['employeeid', 'reviewdate'], inplace=True)

# Add Previous Manager Rating and Days Since Last Review
df_fact['previous_manager_rating'] = df_fact.groupby('employeeid')['managerrating'].shift(1)
df_fact['days_since_last_review'] = df_fact.groupby('employeeid')['reviewdate'].diff().dt.days

# --- 4. Rename Columns to Match Power BI ---
df_fact.rename(columns={
    'envsatisfaction_text': 'envsatisfaction_text',
    'jobsatisfaction_text': 'satisfaction_text_job',
    'relationshipsatisfaction_text': 'satisfaction_text_rel',
    'selfrating_text': 'rating_text_self',
    'managerrating_text': 'managerrating_text'
}, inplace=True)

# Add Work Life Balance Text column
df_fact = pd.merge(
    df_fact,
    df_satisfied_lookup.rename(columns={'satisfaction_text': 'worklifebalance_text'}),
    left_on='worklifebalance',
    right_on='id',
    how='left'
)
# --- 5. Export Performance Fact Table ---
df_fact.to_csv('../data/clean/performance_fact.csv', index=False)
print("âœ… Exported: performance_fact.csv")


# --- 6. Create the Enriched Employee Master Table (employee_master.csv) ---

print("Starting: Creating enriched Employee Master Table...")

df_master = employeeDF.copy()

# Enrich Education
df_master = pd.merge(
    df_master,
    education_levelDF.rename(columns={
        'educationlevelid': 'education',
        'educationlevel': 'education_level_text'
    }),
    on='education',
    how='left'
)



# Drop old numerical ID and Export
df_master.drop(columns=['education'], inplace=True, errors='ignore')
df_master.to_csv('../data/clean/employee_master.csv', index=False)
print("âœ… Exported: employee_master.csv")
```

Aggregation and ML Data Preparation

```python

# --- 4. Aggregate Performance Metrics for ML ---

print("Starting: Creating ML Data Table via Aggregation...")

# df_fact is already sorted from Part 1.

performance_agg = df_fact.groupby('employeeid').agg(
    # Existing Aggregations
    avg_manager_rating=pd.NamedAgg(column='managerrating', aggfunc='mean'),
    total_trainings_taken=pd.NamedAgg(column='trainingopportunitiestaken', aggfunc='sum'),
    num_reviews=pd.NamedAgg(column='employeeid', aggfunc='count'),
    latest_job_sat=pd.NamedAgg(column='jobsatisfaction', aggfunc=lambda x: x.iloc[-1]),
    
    # --- AGGREGATIONS OF NEW FEATURES ---
    # Average change in rating 
    avg_rating_change=pd.NamedAgg(column='previous_manager_rating', aggfunc=lambda x: (x.iloc[-1] - x.iloc[0]) if len(x)>1 else 0),
    # The last recorded value for previous_manager_rating
    previous_rating=pd.NamedAgg(column='previous_manager_rating', aggfunc=lambda x: x.iloc[-1] if len(x)>1 else np.nan),
    # Average time between reviews
    avg_days_between_reviews=pd.NamedAgg(column='days_since_last_review', aggfunc='mean')
    
).reset_index()


# --- 5. Merge Aggregated Performance onto the Employee Master ---
df_ml = pd.merge(
    df_master.copy(), 
    performance_agg,
    on='employeeid',
    how='left'
)

# --- 6. Feature Engineering for ML ---
# Target Variable: Attrition (Yes=1, No=0)
df_ml['attrition_binary'] = df_ml['attrition'].map({'Yes': 1, 'No': 0})
df_ml.drop(columns=['attrition'], inplace=True, errors='ignore')

# OverTime (Yes=1, No=0)
df_ml['overtime_binary'] = df_ml['overtime'].map({'Yes': 1, 'No': 0})
df_ml.drop(columns=['overtime'], inplace=True, errors='ignore')

# Recalculate Tenure/Progression
df_ml['tenure_to_role_ratio'] = df_ml['yearsinmostrecentrole'] / (df_ml['yearsatcompany'] + 0.001)

# Handle NaNs created by new features
df_ml['previous_rating'].fillna(df_ml['avg_manager_rating'], inplace=True)
df_ml['avg_days_between_reviews'].fillna(0, inplace=True)

# Export ML Data Table
df_ml.to_csv('../data/clean/ml_data_for_prediction.csv', index=False)
print("âœ… Exported: ml_data_for_prediction.csv (Ready for Scikit-learn)")
```
