```python
# import sys, os
# sys.path.append(os.path.abspath('../scripts'))
import pandas as pd
# from kpi_functions import *
employeeDF = pd.read_csv('../data/clean/employee_clean.csv')
bigDF = pd.read_csv('../data/clean/bigTable.csv')
performanceDF = pd.read_csv('../data/clean/performance_clean.csv')
```

```python
# gender_distribution
def gender_distribution(df, show_output=True):
    gender_dist = df['gender'].value_counts(normalize=True) * 100
    gender_dist = gender_dist.round(2)
    if show_output:
        print("\n Gender Distribution (%):")
        for key, val in gender_dist.items():
            print(f'{key} : {val} %')
    return gender_dist
```

```python
gender_dist = gender_distribution(employeeDF, True)
print(gender_dist)
```

```python
# Age Distibution
def age_statistics(df, show_output=True):
    """
    Calculates mean, median, and range of employee ages.

    Args:
        df (pd.DataFrame): Employee dataframe.
        show_output (bool): Whether to print the summary.

    Returns:
        dict: Contains 'mean', 'median', and 'range' of Age.
    """
    stats = {
        'mean': df['age'].mean(),
        'median': df['age'].median(),
        'range': df['age'].max() - df['age'].min()
    }
    if show_output:
        print("\nðŸ“Š Age Statistics:")
        print(f"Average Age: {stats['mean']:.2f}")
        print(f"Median Age: {stats['median']}")
        print(f"Age Range: {stats['range']}")
    return stats
```

```python
age_statistics(employeeDF, True)
```

```python
# EDUCATION LEVEL DISTRIBUTION
def education_distribution(df, show_output=True):
    """
    Calculates the number of employees per education level.
    Requires the merged dataset (bigDF) with EducationLevel column.

    Args:
        df (pd.DataFrame): Merged dataset containing 'EducationLevel'.
        show_output (bool): Whether to print the summary.

    Returns:
        pd.Series: Employee counts by EducationLevel.
    """
    edu_dist = df['educationlevel'].value_counts()
    if show_output:
        print("\nðŸŽ“ Education Level Distribution:")
        print(edu_dist)
    return edu_dist
```

```python
education_distribution(bigDF, True)
```

```python
# DEPARTMENT DISTRIBUTION
# -------------------------------------------------------------
def department_distribution(df, show_output=True):
    """
    Counts number of employees in each department.

    Args:
        df (pd.DataFrame): Employee dataframe.
        show_output (bool): Whether to print the summary.

    Returns:
        pd.Series: Employee counts by Department.
    """
    dept_dist = df['department'].value_counts()
    if show_output:
        print("\nðŸ¢ Department Distribution:")
        print(dept_dist)
    return dept_dist
```

```python
department_distribution(employeeDF, True)
```

```python
# TENURE STATISTICS
# -------------------------------------------------------------
def tenure_statistics(df, show_output=True):
    """
    Calculates mean, median, and range of YearsAtCompany.

    Args:
        df (pd.DataFrame): Employee dataframe.
        show_output (bool): Whether to print the summary.

    Returns:
        dict: Contains 'mean', 'median', and 'range' for tenure.
    """
    stats = {
        'mean': df['yearsatcompany'].mean(),
        'median': df['yearsatcompany'].median(),
        'range': df['yearsatcompany'].max() - df['yearsatcompany'].min()
    }
    if show_output:
        print("\nTenure Statistics:")
        print(f"Average Years at Company: {stats['mean']:.2f}")
        print(f"Median Years: {stats['median']}")
        print(f"Range (Max - Min): {stats['range']}")
    return stats
```

```python
tenure_statistics(employeeDF, True)
```

```python
# 2.1 Average Performance Rating (Self + Manager)
def average_performance_rating(df, show_output=True):
    avg_self = df['selfrating'].mean()
    avg_manager = df['managerrating'].mean()

    if show_output:
        print("\nAverage Performance Rating:")
        print(f"Self Rating Average: {avg_self:.2f}")
        print(f"Manager Rating Average: {avg_manager:.2f}")

    return avg_self, avg_manager
```

```python

    # 2.3 Satisfaction and Engagement
    # Calculates average satisfaction levels and distributions.
    
    # Parameters:
    #     df (pd.DataFrame): DataFrame containing satisfaction columns
    #     show_output (bool): Print results if True
    
    # Returns:
    #     dict: A dictionary containing averages and distributions


def satisfaction_engagement(df, show_output=True):
    satisfaction_cols = [
        'environmentsatisfaction',
        'jobsatisfaction',
        'relationshipsatisfaction'
    ]
    
    results = {}

    for col in satisfaction_cols:
        avg_score = df[col].mean()
        distribution = df[col].value_counts(normalize=True).sort_index() * 100
        
        results[col] = {
            'average': round(avg_score, 2),
            'distribution': distribution.round(2).to_dict()
        }

    if show_output:
        print("\nðŸ˜Š Employee Satisfaction & Engagement KPIs:")
        for col, stats in results.items():
            print(f"\nðŸ”¹ {col}:")
            print(f"   Avg Score: {stats['average']}")
            print(f"   Distribution (%): {stats['distribution']}")

    return results
```

```python
satisfaction_engagement(bigDF, True)
```

```python
# 2.4 Training Opportunities
# KPI: Training Utilization
# Purpose:
#     Measures how effectively employees take advantage 
#     of available training opportunities.
#
# Logic:
#     training_utilization = (TrainingTaken / TrainingOffered) * 100
#
# Why it matters:
#     Low utilization may indicate:
#       - low engagement
#       - workload prevents learning
#       - poor training scheduling
#       - lack of support from managers
#     High utilization shows a strong development culture.
#
# Output:
#     New column: 'training_utilization'
#     Returned: EmployeeID and utilization percentage

def training_utilization_kpi(df, show_output=True):
    utilization = []

    for i, row in df.iterrows():
        if row['trainingopportunitieswithinyear'] > 0:
            percent = (row['trainingopportunitiestaken'] /
                       row['trainingopportunitieswithinyear']) * 100
        else:
            percent = 0
        utilization.append(percent)

    df['training_utilization'] = utilization

    if show_output:
        print("\nTraining Utilization (%):")
        print(df['training_utilization'].describe())

    return df[['employeeid', 'training_utilization']]
```

```python
# 2.5 Work-Life Balance
# KPI: Work-Life Balance Score
# Purpose:
#     Measures how well employees are balancing work with personal life.
#
# Logic:
#     Average rating of 'WorkLifeBalance' column.
#     Higher score = healthier work-life balance.
#
# Business Value:
#     Helps HR detect burnouts, workload issues,
#     and areas where wellness programs are needed.
#
# Output:
#     Overall average score and distribution visualization.
# -------------------------------------------------------
def work_life_balance_kpi(df, show_output=True):
    if 'worklifebalance' not in df.columns:
        print("Column 'worklifebalance' not found in the dataset.")
        return None

    avg_wlb = df['worklifebalance'].mean()

    if show_output:
        print("\nðŸŒ± Work-Life Balance KPI:")
        print(f"Average Score: {avg_wlb:.2f}")
        print(df['worklifebalance'].describe())

    return avg_wlb
```

```python
work_life_balance_kpi(bigDF, True)
```

```python
# 3.1 Attrition Rate
# KPI: Attrition Distribution (Stayed vs Left)
# Purpose:
#     Shows the percentage of employees who stayed vs left.
#
# Business Value:
#     Gives HR a clear view of retention health.
#
# Output:
#     Series with two percentages: Yes (left) / No (stayed)
# -------------------------------------------------------
def attrition_distribution_kpi(df, show_output=True):

    if 'attrition' not in df.columns:
        print("attrition' column missing from dataset")
        return None

    distribution = df['attrition'].value_counts(normalize=True) * 100
    distribution = distribution.round(2).astype(str) + '%'
    if show_output:
        print("\nAttrition Distribution (%):")
        print(distribution)

    return distribution
```

```python
attrition_distribution_kpi(employeeDF, True)
```

```python
# 3.2 Attrition by Department
# KPI: Attrition by Department
# Purpose:
#     Measures turnover rate per department to identify
#     areas needing HR attention and retention efforts.
#
# Output:
#     DataFrame with Department and Attrition %
# -------------------------------------------------------
def attrition_by_department_kpi(df, show_output=True):
    if 'attrition' not in df.columns or 'department' not in df.columns:
        print("Required columns missing: 'attrition', 'department'")
        return None
    
    # Calculate attrition rate per department
    dept_attrition = df.groupby('department')['attrition'].apply(
        lambda x: (x == 'Yes').mean() * 100
    )

    dept_attrition = dept_attrition.round(2)

    if show_output:
        print("\nAttrition by Department (%):")
        print(dept_attrition)

    return dept_attrition
```
